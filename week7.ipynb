```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Solve linear system
A1 = np.array([
    [27, -10, 4, -29],
    [-16, 5, -2, 18],
    [-17, 4, -2, 20],
    [-7, 2, -1, 8]
])
b = np.array([1, -1, 0, 1])
x = np.linalg.solve(A1, b)
print("Problem 1 Solution:")
print(f"x1 = {x[0]:.0f}, x2 = {x[1]:.0f}, x3 = {x[2]:.0f}, x4 = {x[3]:.0f}")

# Problem 2a: Compute T(x)
A2 = np.array([
    [np.sqrt(2)/2, -np.sqrt(2)/2],
    [np.sqrt(2)/2, np.sqrt(2)/2]
])
v = np.array([2, -2])
Tv = A2 @ v
print("\nProblem 2(a):")
print(f"T([2, -2]) = [{Tv[0]:.3f}, {Tv[1]:.3f}]")

# Problem 2b: Visualization
plt.figure(figsize=(10, 5))

# Original vectors
plt.subplot(121)
plt.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='r', width=0.015)
plt.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='b', width=0.015)
plt.quiver(0, 0, 2, -2, angles='xy', scale_units='xy', scale=1, color='g', width=0.015)
plt.title("Original Vectors")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()

# Transformed vectors
plt.subplot(122)
plt.quiver(0, 0, A2[0,0], A2[1,0], angles='xy', scale_units='xy', scale=1, color='r', width=0.015)
plt.quiver(0, 0, A2[0,1], A2[1,1], angles='xy', scale_units='xy', scale=1, color='b', width=0.015)
plt.quiver(0, 0, Tv[0], Tv[1], angles='xy', scale_units='xy', scale=1, color='g', width=0.015)
plt.title("Transformed Vectors")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()

plt.tight_layout()
plt.savefig('transformation.png', dpi=100)
print("\nProblem 2(b): Visualization saved as 'transformation.png'")

# Problem 2c: Geometric description
print("\nProblem 2(c):")
print("The transformation is a counterclockwise rotation by 45 degrees.")
```
